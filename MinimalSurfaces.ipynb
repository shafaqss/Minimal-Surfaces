```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
u = np.linspace(0, 2*np.pi, 32)
v = np.linspace(0, 2*np.pi, 32)
U, V = np.meshgrid(u,v)
a, r = 1.0, 0.25;

X = (a + r*np.cos(U))*np.cos(V)
Y = (a + r*np.cos(U))*np.sin(V)
Z = r*np.sin(U)

fig = plt.figure()
plt.figure(figsize=(8, 8), dpi=100)

ax = plt.axes(projection='3d')
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('Torus');
#plt.axis('off')
plt.show()
```

```python

u = np.linspace(0, 2*np.pi, 50)
v = np.linspace(-1.2, 1.2, 50)
U, V = np.meshgrid(u,v)
a = 2.

X = a * np.cosh(V)*np.cos(U)
Y = a * np.cosh(V)*np.sin(U)
Z = a*V

fig = plt.figure()
plt.figure(figsize=(8, 8), dpi= 100)

ax = plt.axes(projection='3d')
#ax.set_xlim3d(-1, 1)
#ax.set_ylim3d(-1, 1)
#ax.set_zlim3d(-1, 1)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='twilight_shifted') #shade=True, antialiased=False)
#ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='twilight_shifted',antialiased=False)
ax.set_title('surface catenoid');
#plt.axis('off')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('wireframe');
```

```python
u = np.linspace(-10, 10, 50)
v = np.linspace(-10, 10, 50)
U, V = np.meshgrid(u,v)

X = U - (U**3)/3 + U*(V**2)
Y = V - (V**3)/3 + V*(U**2)
Z = U**2 - V**2

fig = plt.figure()
ax = plt.axes(projection='3d')
#ax.set_xlim3d(-1, 1)
#ax.set_ylim3d(-1, 1)
#ax.set_zlim3d(-1, 1)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='twilight_shifted')
ax.set_title('surface enneper');
```

```python

u1 = np.linspace(np.pi/2+0.001, 3*np.pi/2-0.001, 50)
v1 = np.linspace(np.pi/2+0.001, 3*np.pi/2-0.001, 50)


#u1 = np.linspace(-np.pi/2+0.001, np.pi/2-0.001, 50)
#v1 = np.linspace(-np.pi/2+0.001, np.pi/2-0.001, 50)
U1, V1 = np.meshgrid(u1,v1)

X1 = U1 
Y1 = V1
Z1 = np.log(np.cos(V1)/np.cos(U1))

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X1, Y1, Z1, rstride=1, cstride=1,
                cmap='twilight_shifted', edgecolor='none')
ax.set_title('surface scherk');
plt.show()
```
